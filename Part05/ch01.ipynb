## 2유형

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/Bank_Personal_Loan_Modeling.csv')
df
```

```python
df.info()
```

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
```

```python
df = df.drop(columns={'ID','ZIP Code'},axis=1)
df.info()
```

```python
X = df.drop(['Personal Loan'],axis=1)
y = df['Personal Loan']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 1234)
```

```python
import sklearn.preprocessing as preprocessing
preprocessor = preprocessing.Normalizer()
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.fit_transform(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1,25)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(X_train,y_train)

    training_accuracy.append(clf.score(X_train,y_train))

    test_accuracy.append(clf.score(X_test,y_test))
```

```python
test_accuracy
```

```python
np.max(test_accuracy)
```

## 3유형

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv')
df
```

```python
# T-Test

import numpy as np
from scipy import stats

alpha = 0.05

alternative = 'two-sided'

t_statistic,p_value = stats.ttest_1samp(df['Temp'],77,alternative = alternative)

confidence_interval = stats.t.interval(1-alpha,len(df['Temp'])-1,loc = np.mean(df['Temp']),scale = stats.sem(df['Temp']))

print(np.mean(df['Temp']))
print('t-통계량:',t_statistic)
print('p-값:',p_value)
print('95% 신뢰구간:',confidence_interval)

```

```python
ploan_processed = ploan.dropna().drop(['ID','ZIP Code'],axis = 1,inplace=False)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 1234)

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
import sklearn.preprocessing as preprocessing

preprocessor = preprocessing.Normalizer()
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.fit_transform(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1,25)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)

    training_accuracy.append(clf.score(X_train,y_train))

    test_accuracy.append(clf.score(X_test,y_test))
```
