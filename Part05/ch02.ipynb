## 2유형

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/Parkinsons.csv')
df
```

```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
from sklearn.model_selection import train_test_split
import sklearn.preprocessing as preprocessing
from sklearn import metrics
from sklearn.metrics import f1_score
```

```python
df.head(10)
```

```python
# name 칼럼 제거
dat_processing = df.drop(['name'],axis = 1,inplace=False)
```

```python
dat_processing_norm = preprocessing.minmax_scale(dat_processing)
dat_processed = pd.DataFrame(dat_processing_norm)
dat_processed.columns = dat_processing.columns
```

```python
# 상수항 추가
dat_processed = sm.add_constant(dat_processed,has_constant = 'add')
dat_processed.head()
```

```python
feature_columns = list(dat_processed.columns.difference(['status']))

X = dat_processed[feature_columns]
y = dat_processed['status'].astype('category')
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,y,stratify=y,test_size=0.1,random_state = 2017010500)
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
model = sm.Logit(train_y,train_x)
```

```python
results = model.fit(method = 'bfgs',maxiter = 1000)
```

```python
results.summary()
```

```python
# cut off 정의
def cut_off(y,threshold):
    Y = y.copy()
    Y[Y>threshold] = 1
    Y[Y<=threshold] = 0
    return(Y.astype(int))
```

```python
test_y_pred_prob = results.predict(test_x)
test_y_pred = cut_off(test_y_pred_prob,0.8)
```

```python
f1_score(test_y,test_y_pred)
```

# 3유형

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/HR-Employee-Attrition.csv')
```

```python
df
```

```python
df[['MaritalStatus','OverTime']]
```

```python
# 결혼 유무와 초과근로 간의 분할표
table = pd.crosstab(df['OverTime'],df['MaritalStatus'])
print(table)
```

```python
# X1 : 초과근로자에 대한 결혼 여부
# X2 : 초과근로 하지 않은 자의 결혼 여부

X1 = table.loc['Yes',:]
X2 = table.loc['No',:]
print(X1)
print(X2)
```

```python
X1_Married = X1['Married']
X2_Married = X2['Married']

print(int(abs(X1_Married - X2_Married)))
```

```python
# 2 
result = stats.chi2_contingency([X1,X2])
print(round(result.statistic,2))
```

```python
p_value = result.pvalue
print(round(p_value,4))

if p_value<0.05:
    print('기각')
else:
    print('채택')
```
